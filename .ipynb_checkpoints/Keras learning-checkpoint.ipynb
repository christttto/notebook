# 04 Create an artificial neural network with Keras


https://www.youtube.com/watch?v=Boo6SmgmHuM&list=PLZbbT5o_s2xrwRnXk_yCPtnqqo4_u2YGL&index=4

```python
import keras
from keras import backend as TF
```

"from keras import backend as K" was changed to  
"from keras import backend as TF" because I am using tensorflow as my backend

```python
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(16, input_shape=(1,),activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
```

passing in the layers into the model.  
with dense layer.  
Dense expects the number of neuron in the beginning, so will be 16.  
the input shape will be with "input_shape" that has 1 demensional.  
activation that I want to use can be put in optionally.  
  
the last layer is output layer.  
the number of cases. yes reaction, no reaction

```python
model.summary()
```

```python
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

learning rate, loss calcuation, metrics is a way to measure how well performed

valid_set = [(sample, label), ..., (sample,label)]  
a way to provide seperate testing stuff  
but with validation split, it can be traning on a seperated stuff

```python
model.fit(scaled_train_samples, train_labels, validation_split=0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

training model using fit function.  
batch size is how many samples to be grouped together.
shuffle will be true, will give each epoch with different order.
